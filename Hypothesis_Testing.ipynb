# Hypothesis Testing

## import dependencies

```python
import math
from scipy.stats import norm
```

```python
sample_mean = 3050
population_mean = 1000 + 5 * 600
pupulation_std = 5 * 25
n = 25
alpha = 0.05
```

## critical value

```python
critical_value = norm.ppf(1 - alpha)
critical_value
```

## test statistic (because we know the population mean)

```python
z_score = (sample_mean - population_mean)/ (pupulation_std/math.sqrt(n))
z_score
```

## conclusion

```python
print(f'z_score({z_score}) < critical value({critical_value})')
print('So, that we failed to reject the null hypothesis which is the mean of weekly operating cost \nis not greater than theoretical weekly operating cost model($1,000 + $5X)')
```
